###### First, we'll read in the data and take a look at a few entries to see what we've got:

```python
cfpb_complaints = pd.read_csv("complaints_dec_2014.csv")
```

```python
cfpb_complaints.head()
```

###### Sorting complaints by product type:

```python
complaints_by_product = cfpb_complaints.Product.value_counts()
print(complaints_by_product)
complaints_by_product.plot(kind='bar', figsize=(18,5))
plt.show()
```

```python
complaints_by_product.plot(kind='pie', figsize = (9,9))
plt.show()
```

This raw data is fairly useful in determining what sorts of product complaints the bureau receives.  The charts give an easy visualization of this.  Basically half the complaints are related to either mortgages or debt collection.

###### Sorting complaints by company:

```python
complaints_by_company = cfpb_complaints.Company.value_counts()
top10_complaints_by_company = complaints_by_company[:10]
print(top10_complaints_by_company)
top10_complaints_by_company.plot(kind='bar', figsize=(18,5))
plt.show()
```

We can see that the big banks and credit rating services are the main companies.  In my opinion, this data isn't tremendously useful without the ability to see how these numbers compare to the total number of users/customers of these companies.  If we had that data, we would be able to see which of these companies cause higher/lower relative complaint rates among their users/customers.

###### Sorting complaints by type of company response:

```python
complaints_by_company_response = cfpb_complaints["Company response"].value_counts()
print(complaints_by_company_response)
complaints_by_company_response.plot(kind='bar', figsize=(18,5))
plt.show()
```

```python
complaints_by_company_response.plot(kind='pie', figsize = (6,6))
plt.show()
```

###### Let's take a look at the days of the week

```python
cfpb_complaints['dates'] = pd.to_datetime(cfpb_complaints.pop('Date received'), format="%m/%d/%Y")
```

```python
cfpb_complaints.head()
```

```python
complaints_by_day = cfpb_complaints.dates.dt.weekday
# complaints_by_day
```

```python
total_complaints_by_day = complaints_by_day.value_counts()
print(total_complaints_by_day)
```

```python
total_complaints_by_day = total_complaints_by_day.sort_index()
total_complaints_by_day
```

```python
total_complaints_by_day.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
print(total_complaints_by_day)
```

###### Here I cheat by looking up in a calendar, how many of each weekday there was in December 2014

```python
# 3 * Mon, Tues, Wed   4* rest of the week

avg_complaints_by_day = total_complaints_by_day.copy()
avg_complaints_by_day[['Monday', 'Tuesday', 'Wednesday']] = avg_complaints_by_day[['Monday', 'Tuesday', 'Wednesday']]/5
avg_complaints_by_day[['Thursday', 'Friday', 'Saturday', 'Sunday']] = (
avg_complaints_by_day[['Thursday', 'Friday', 'Saturday', 'Sunday']]/4)

avg_complaints_by_day
```

```python
avg_complaints_by_day.plot(kind='bar', figsize=(18,5))
plt.show()
```

```python
avg_complaints_by_day.plot(kind='pie', figsize=(6,6))
plt.show()
```

Complaining to the CFPB is weekday work!  The most common days of the week for complaints are Monday and Tuesday.  Complaints trail off as the week goes on and significantly dip over the weekend.
